```python
import opendatasets as od
import os 
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
dir='/kaggle/input/dogs-cats-images/dataset/'
```

```python
categories = ['cats', 'dogs']
test_dir = '/kaggle/input/dogs-cats-images/dataset/test_set/'
train_dir = '/kaggle/input/dogs-cats-images/dataset/training_set/'

data = []

for i in categories:
    # get path for each directory category by combining the category name and dir path
    path = os.path.join(train_dir, i)
    # get label for all images in directory
    label = categories.index(i)

    for img in os.listdir(path):
        # get path for each image by combining the image name and dir path
        imgpath = os.path.join(path, img)
        # 0 indicates that the image should be read in grayscale
        pet_img = cv2.imread(imgpath, 0)
        # resize the image
        pet_img = cv2.resize(pet_img, (50, 50))
        # convert from 2-D array into 1-D
        image = np.array(pet_img).flatten()
        # append data for each image
        data.append([image, label])

print(len(data))
```

```python
print(data[11])
```

```python
features=[]
labels=[]

for feature,label in data:
    features.append(feature)
    labels.append(label)
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.20,random_state=77)
```

```python
model=svm.SVC(C=1,kernel='poly',gamma='auto')
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
accuracy= model.score(x_test,y_test)
accuracy 
```
